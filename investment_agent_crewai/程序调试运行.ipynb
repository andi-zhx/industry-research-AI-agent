````python
# Pythonä»£ç ç”Ÿæˆppt
from pptx import Presentation
from pptx.util import Inches, Pt
from pptx.dml.color import RGBColor
from pptx.enum.text import PP_ALIGN
````

````python
# --- é…ç½®é…è‰²æ–¹æ¡ˆ (é‡‘èè“é£æ ¼) ---
DARK_BLUE = RGBColor(30, 58, 138)  # ä¸»æ ‡é¢˜è‰²
LIGHT_BLUE = RGBColor(59, 130, 246) # å¼ºè°ƒè‰²
GRAY_TEXT = RGBColor(80, 80, 80)    # æ­£æ–‡è‰²

def create_presentation():
    prs = Presentation()

    # --- è¾…åŠ©å‡½æ•°ï¼šè®¾ç½®æ–‡æœ¬æ ¼å¼ ---
    def set_font(run, size, bold=False, color=GRAY_TEXT):
        run.font.name = 'Arial' # æˆ– 'Microsoft YaHei'
        run.font.size = Pt(size)
        run.font.bold = bold
        run.font.color.rgb = color

    # --- è¾…åŠ©å‡½æ•°ï¼šåˆ›å»ºæ ‡é¢˜å¹»ç¯ç‰‡ ---
    def add_title_slide(title_text, subtitle_text):
        slide = prs.slides.add_slide(prs.slide_layouts[0])
        title = slide.shapes.title
        subtitle = slide.placeholders[1]

        title.text = title_text
        set_font(title.text_frame.paragraphs[0].runs[0], 44, True, DARK_BLUE)
        
        subtitle.text = subtitle_text
        set_font(subtitle.text_frame.paragraphs[0].runs[0], 24, False, LIGHT_BLUE)

    # --- è¾…åŠ©å‡½æ•°ï¼šåˆ›å»ºå†…å®¹å¹»ç¯ç‰‡ ---
    def add_content_slide(title_text, content_list):
        slide = prs.slides.add_slide(prs.slide_layouts[1])
        title = slide.shapes.title
        body = slide.placeholders[1]

        title.text = title_text
        set_font(title.text_frame.paragraphs[0].runs[0], 32, True, DARK_BLUE)

        tf = body.text_frame
        for i, item in enumerate(content_list):
            if i == 0:
                p = tf.paragraphs[0]
            else:
                p = tf.add_paragraph()
            
            p.text = item
            p.space_after = Pt(14)
            set_font(p.runs[0], 20, False, GRAY_TEXT)

    # --- è¾…åŠ©å‡½æ•°ï¼šåˆ›å»ºå¯¹æ¯”è¡¨æ ¼å¹»ç¯ç‰‡ (ç«äº‰ä¼˜åŠ¿) ---
    def add_table_slide(title_text, data):
        slide = prs.slides.add_slide(prs.slide_layouts[5]) # Title Only layout
        title = slide.shapes.title
        title.text = title_text
        set_font(title.text_frame.paragraphs[0].runs[0], 32, True, DARK_BLUE)

        # å®šä¹‰è¡¨æ ¼ä½ç½®å’Œå¤§å°
        rows = len(data)
        cols = len(data[0])
        left = Inches(0.5)
        top = Inches(1.5)
        width = Inches(9.0)
        height = Inches(0.8 * rows)

        table = slide.shapes.add_table(rows, cols, left, top, width, height).table

        # è®¾ç½®åˆ—å®½
        table.columns[0].width = Inches(2.0)
        table.columns[1].width = Inches(2.3)
        table.columns[2].width = Inches(2.3)
        table.columns[3].width = Inches(2.4)

        for i in range(rows):
            for j in range(cols):
                cell = table.cell(i, j)
                cell.text = data[i][j]
                
                # æ ¼å¼åŒ–å•å…ƒæ ¼æ–‡æœ¬
                run = cell.text_frame.paragraphs[0].runs[0]
                run.font.size = Pt(14)
                run.font.name = 'Arial'
                
                # è¡¨å¤´åŠ ç²—å˜è‰²
                if i == 0:
                    run.font.bold = True
                    run.font.color.rgb = DARK_BLUE
                    cell.fill.solid()
                    cell.fill.fore_color.rgb = RGBColor(220, 230, 241)
                # æˆ‘ä»¬çš„äº§å“åˆ—åŠ ç²—
                elif j == 3:
                    run.font.bold = True
                    run.font.color.rgb = DARK_BLUE

    # ================= å¹»ç¯ç‰‡å†…å®¹ç”Ÿæˆ =================

    # Slide 1: å°é¢
    add_title_slide(
        "FinSight AI Pro", 
        "ä¸‹ä¸€ä»£æ™ºèƒ½æŠ•ç ”ä¸å†³ç­–ç»ˆç«¯\n\nè®© AI æˆä¸ºæ‚¨çš„é¦–å¸­åˆ†æå¸ˆï¼Œé‡å¡‘ä¸€çº§å¸‚åœºæŠ•ç ”æ•ˆç‡"
    )

    # Slide 2: ç—›ç‚¹
    add_content_slide("ç—›ç‚¹ä¸æœºé‡ï¼šæŠ•ç ”è¡Œä¸šçš„æ•ˆç‡å±æœº", [
        "âŒ æ•°æ®å­¤å²›ä¸ç¢ç‰‡åŒ–ï¼šåˆ†æå¸ˆ 70% æ—¶é—´æµªè´¹åœ¨æœé›†æ•°æ®ã€æ•´ç†çºªè¦ç­‰â€œæ¬ç –â€å·¥ä½œã€‚",
        "âŒ åˆçº§å·¥ä½œé‡å¤ä½æ•ˆï¼šæ’°å†™åŸºç¡€ç ”æŠ¥ã€å½•å…¥è´¢åŠ¡æ¨¡å‹è€—æ—¶æé•¿ä¸”æ˜“å‡ºé”™ã€‚",
        "âŒ é€šç”¨ AI ä¸å¤Ÿä¸“ä¸šï¼šChatGPT ç¼ºä¹é‡‘èå‚ç±»çŸ¥è¯†ï¼Œä¸æ‡‚å››çº§äº§ä¸šé“¾ï¼Œæ˜“äº§ç”Ÿå¹»è§‰ã€‚",
        "âŒ éœ€æ±‚åƒäººåƒé¢ï¼šæ”¿åºœå…³æ³¨æ‹›å•†ï¼ŒVCå…³æ³¨é«˜å¢é•¿ï¼Œä¼ ç»Ÿå·¥å…·æ— æ³•çµæ´»åˆ‡æ¢è§†è§’ã€‚"
    ])

    # Slide 3: è§£å†³æ–¹æ¡ˆ
    add_content_slide("è§£å†³æ–¹æ¡ˆï¼šFinSight AI Pro", [
        "âœ… å…¨æµç¨‹è‡ªåŠ¨åŒ–ï¼šå¤§çº²è§„åˆ’ -> å¼‚æ­¥æœé›† -> é€»è¾‘åˆ†æ -> æŠ¥å‘Šæ’°å†™ -> å®¡æ ¸ï¼Œå…¨é—­ç¯ã€‚",
        "âœ… åƒäººåƒé¢å®šåˆ¶ï¼šæ”¯æŒ Gç«¯(æ”¿åºœ)ã€Bç«¯(VC)ã€é£æ§ç«¯ å¤šè§†è§’åˆ‡æ¢ã€‚",
        "âœ… æé€Ÿå¹¶è¡Œå¤„ç†ï¼šå¼‚æ­¥å¹¶è¡Œæ¶æ„ï¼Œå°† 3 å¤©çš„ç ”æŠ¥å·¥ä½œå‹ç¼©è‡³ 5 åˆ†é’Ÿã€‚",
        "âœ… æ·±åº¦é€»è¾‘ç©¿é€ï¼šå†…ç½®â€œå››çº§äº§ä¸šé“¾â€å›¾è°±ï¼Œç©¿é€å®è§‚èµ›é“è‡³å¾®è§‚æŠ€æœ¯èŠ‚ç‚¹ã€‚"
    ])

    # Slide 4: æ ¸å¿ƒæŠ€æœ¯
    add_content_slide("æ ¸å¿ƒæŠ€æœ¯å£å’ï¼šå·¥ç¨‹åŒ–çš„æ™ºèƒ½ä½“ç³»ç»Ÿ", [
        "ğŸ§  CrewAI å¤šæ™ºèƒ½ä½“æ¶æ„ï¼šPlanner(å¤§è„‘) + Researcher(æœé›†) + Analyst(é€»è¾‘) + Writer(ç¬”æ†)ã€‚",
        "âš¡ å¼‚æ­¥å¹¶è¡Œæœç´¢ï¼šçªç ´ä¸²è¡Œç“¶é¢ˆï¼Œå¤šçº¿ç¨‹å¹¶å‘è·å–è´¢åŠ¡ã€æ”¿ç­–ã€è¡Œä¸šæ•°æ®ã€‚",
        "ğŸ“š RAG + å®æ—¶è”ç½‘ï¼šæœ¬åœ°çŸ¥è¯†åº“(ç§æœ‰æ•°æ®) ä¸ å®æ—¶è”ç½‘(æœ€æ–°æ•°æ®) åŒå¼•æ“é©±åŠ¨ã€‚",
        "ğŸ“Š ç»“æ„åŒ–è¾“å‡ºï¼šå¼ºåˆ¶ç”Ÿæˆ Markdown è¡¨æ ¼ä¸å¯è§†åŒ–å›¾è¡¨ï¼Œéå•çº¯æ–‡æœ¬å †ç Œã€‚"
    ])

    # Slide 5: äº§å“åŠŸèƒ½çŸ©é˜µ
    add_content_slide("äº§å“åŠŸèƒ½çŸ©é˜µï¼šé‡‘èæŠ•ç ”çš„â€œç‘å£«å†›åˆ€â€", [
        "ğŸ“Š è¡Œä¸šæ·±åº¦ç ”ç©¶ï¼šå››çº§äº§ä¸šé“¾ç©¿é€ï¼Œå¤šè§†è§’è‡ªåŠ¨ç”Ÿæˆæ·±åº¦ç ”æŠ¥ã€‚",
        "âš–ï¸ æ™ºèƒ½å°½èŒè°ƒæŸ¥ï¼šGrowth vs Risk æ¡†æ¶ï¼Œè‡ªåŠ¨æ‰«æè´¢åŠ¡ç²‰é¥°ä¸æ³•å¾‹çº¢çº¿ã€‚",
        "ğŸ’° è´¢åŠ¡ä¼°å€¼å»ºæ¨¡ï¼šè‡ªåŠ¨è¯»å– Pitchbookï¼Œä¸€é”®ç”Ÿæˆ DCF æ¨¡å‹ä¸ Football Fieldã€‚",
        "ğŸ“ æ™ºèƒ½ä¼šè®®çºªè¦ï¼šéŸ³è§†é¢‘è½¬å½• + è¯´è¯äººè¯†åˆ« + Action Item æå–ã€‚",
        "ğŸ“‘ BP/è´¢æŠ¥è§£è¯»ï¼šè‡ªåŠ¨æå– SWOTï¼Œè®¡ç®— KPIï¼Œé£é™©é¢„è­¦ã€‚"
    ])

    # Slide 6: ç«äº‰ä¼˜åŠ¿ (è¡¨æ ¼)
    table_data = [
        ["ç»´åº¦", "ä¼ ç»ŸæŠ•ç ”å·¥å…·", "é€šç”¨å¤§æ¨¡å‹", "FinSight AI Pro"],
        ["ä¸“ä¸šåº¦", "ä»…æä¾›æ•°æ®ï¼Œæ— è§‚ç‚¹", "æ˜“èƒ¡è¯´ï¼Œæ•°æ®æ»å", "æ•°æ®å®æ—¶+é€»è¾‘ä¸¥å¯†+è‡ªåŠ¨åˆ¶è¡¨"],
        ["æµç¨‹", "çº¯äººå·¥æ“ä½œ", "ç®€å•é—®ç­”äº¤äº’", "Agentic Workflow (è‡ªä¸»æ‰§è¡Œ)"],
        ["å®šåˆ¶åŒ–", "æ— ", "éœ€å¤æ‚Promptå·¥ç¨‹", "å†…ç½®ä¸“å®¶Promptï¼Œä¸€é”®åˆ‡æ¢"],
        ["æ•ˆç‡", "3å¤© / ç¯‡", "éœ€åå¤å¯¹è¯è°ƒæ•´", "5åˆ†é’Ÿ / ç¯‡ (å¼‚æ­¥å¹¶è¡Œ)"]
    ]
    add_table_slide("ç«äº‰ä¼˜åŠ¿ï¼šWhy Us?", table_data)

    # Slide 7: å•†ä¸šæ¨¡å¼
    add_content_slide("å•†ä¸šæ¨¡å¼ä¸å¸‚åœº", [
        "1. SaaS è®¢é˜…æœåŠ¡ï¼šé¢å‘ä¸­å°å‹æœºæ„ï¼ŒæŒ‰æœˆ/å¹´ä»˜è´¹ã€‚",
        "2. API / SDK é›†æˆï¼šé›†æˆè‡³ Windã€åŒèŠ±é¡ºæˆ–åˆ¸å•†å†…éƒ¨ç³»ç»Ÿï¼ŒæŒ‰è°ƒç”¨é‡æ”¶è´¹ã€‚",
        "3. ç§æœ‰åŒ–éƒ¨ç½²ï¼šé¢å‘å›½èµ„/å¤§å‹åŸºé‡‘ï¼Œæœ¬åœ°åŒ–éƒ¨ç½²ï¼Œç¡®ä¿æ•°æ®ä¸å‡ºåŸŸã€‚"
    ])

    # Slide 8: ç»“å°¾
    add_title_slide(
        "FinSight AI Pro",
        "å¯»æ±‚èèµ„ / åˆä½œä¼™ä¼´\n\nè”ç³»æˆ‘ä»¬: [æ‚¨çš„åå­—/è”ç³»æ–¹å¼]"
    )

    # ä¿å­˜æ–‡ä»¶
    output_filename = "FinSight_AI_Pro_BP.pptx"
    prs.save(output_filename)
    print(f"âœ… PPT å·²ç”Ÿæˆï¼š{output_filename}")

if __name__ == "__main__":
    create_presentation()
````
